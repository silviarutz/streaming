```python
data = [
    go.Histogram( x = X_train['Year'])
    ]

fig = go.Figure(data)
fig.update_xaxes(title_text='Jahr') #Achsen immer Beschriften
fig.update_yaxes(title_text='Anzahl Filme')
fig.show()
```
