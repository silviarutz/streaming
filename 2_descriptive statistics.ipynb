# Histogramm:
Ausgabe der Variablen (Erscheinungs-)Jahr und Anzahl Serien als Histogramm:

```python
data = [
    go.Histogram( x = X_train['Year'])
    ]

fig = go.Figure(data)
fig.update_xaxes(title_text='Alter der Serie in Jahren') #Achsen immer Beschriften
fig.update_yaxes(title_text='Anzahl Serien')
fig.show()
```

Ausgabe der Variablen (Erscheinungs-)Jahr und Anzahl Serien in Prozent als Histogramm:


```python
data = [
    go.Histogram( x = X_train['Year'],
                  histnorm = 'probability') #berechnet Anteil
    ]

fig = go.Figure(data)
fig.update_xaxes(title_text='Alter der Serie in Jahren')
fig.update_yaxes(title_text='Anzahl Serien in Prozent')
fig.show()
```

Ausgabe der Variablen IMDb (Rating) und Anzahl Serien als Histogramm:


```python
data = [
    go.Histogram( x = X_train['IMDb'])
    ]

fig = go.Figure(data)
fig.update_xaxes(title_text='Rating in IMDb') #Achsen immer Beschriften
fig.update_yaxes(title_text='Anzahl Serien')
fig.show()
```

# Ausgabe Boxplot
Ausgabe der Variable IMDb als Boxplot, um die wichtigsten Quantile des Datensatzes anzuzeigen:
-> Minimum, unteres Quartil, Median, oberes Quartil, Maximum

Wir sehen: Die schlechteste Bewertung lautet 1.1 (Minimum), das untere Quartil liegt bei 4.6, der Median bei 7.3 und die beste IMDb Bewertung lautet 9.6.


```python
data = [
    go.Box( x = X_train['IMDb'])
    ]

fig = go.Figure(data)
fig.update_xaxes(title_text='Rating IMDb')
fig.show()
```
